# Developer Stress Analysis

This notebook replicates the analyses from Lab-1 but uses the `developer_stress.csv` dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the developer stress dataset
df = pd.read_csv('developer_stress.csv')
```

```python
# Quick peek at the data
df.head()
```

```python
# Info and types
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Column names
df.columns
```

```python
# Pairplot (may be slow for many columns)
sns.pairplot(df)
plt.show()
```

```python
# Histogram of the first numeric column (safe default)
num_cols = df.select_dtypes(include='number').columns
if len(num_cols) > 0:
    col = num_cols[0]
    df[col].hist(bins=30, figsize=(8,4))
    plt.title(f'Histogram of {col}')
    plt.show()
else:
    print('No numeric columns to plot.')
```

```python
# Correlation heatmap for numeric features
plt.figure(figsize=(10,7))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, linewidths=2)
plt.title('Numeric feature correlations')
plt.show()
```
